Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (7 points)

Goal_diff_count = GoalsFor - GoalsAgainst

# Solution 4

```python
#convert column data type to numeric
teams['GoalsFor'] = pd.to_numeric(teams['GoalsFor'])
teams['GoalsAgainst'] = pd.to_numeric(teams['GoalsAgainst'])
```

```python
#Below function returns all the teams with their Goal Differences
Goal_diff_count = lambda x,y,z:(x,y-z)
GoalDiff = pd.DataFrame(list(Goal_diff_count(teams['Team'],teams['GoalsFor'],teams['GoalsAgainst'])),index=['Team','GoalDiffCount']).T
print(GoalDiff)
```

```python
#finding team which has maximum goal difference
GoalDiff[GoalDiff["GoalDiffCount"]==GoalDiff["GoalDiffCount"].max()]
```

```python
#finding team which has minimum goal difference
GoalDiff[GoalDiff["GoalDiffCount"]==GoalDiff["GoalDiffCount"].min()]
```

# Solution 5

```python
#convert column data type to numeric
teams['GamesWon'] = pd.to_numeric(teams['GamesWon'])
teams['GamesPlayed'] = pd.to_numeric(teams['GamesPlayed'])
```

```python
#created new column with name “Winning Percent” and appended to teams dataset
teams['Winning Percent'] = (teams['GamesWon'] / teams['GamesPlayed'])*100 

#Replacing any numerical error with 0%
teams.fillna(value=0)

#top 5 teams which has the highest Winning percentage
teams.sort_values(by='Winning Percent',ascending = False).head(5) 
```
